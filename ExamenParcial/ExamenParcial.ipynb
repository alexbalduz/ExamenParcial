```python
#Añadimos columna conversión con 1.
def columna_conversion(df):
    conversion = 1
    df['conversion'] = conversion
```

```python
def columna_reemplazo(df):
    df.loc[df['conversion'].isna(), 'conversion']=0
```

```python
def union_df(df1,df2):
    columna_conversion(df1)
    df_total = pd.concat([df2, df1])
    columna_reemplazo(df_total)
    return df_total
```
